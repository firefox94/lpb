```python
import qrcode
from PIL import Image

# Input data to generate QR code
data = r'https://datapot.vn/top-cac-ham-dax-thuong-dung-trong-power-bi/?srsltid=AfmBOooa7ylNbdm-fCusEke6-dIYiOJYtwkiEIkpPjW55LyjB7_cLYH4'

# Generate qrcode
qr = qrcode.QRCode(version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
    )
qr.add_data(data)
qr.make(fit=True)

# Image
img = qr.make_image(fill='black', back_color='white')

# Save & open img
img.save("qrcode.png")
Image.open("qrcode.png")
```
